### face detection through webcam 

```python
face=r"C:\Documents and Settings\rohan\face detection\haarcascade_frontalface_default (1).xml"

cap=cv2.VideoCapture(0)
ret,frame=cap.read()
while ret:
    ret,frame=cap.read()#if camera is opened then ret  will return  to and frame will return array of captured image
    print(ret)#print true or false
    print(frame)#print array of captured image
    if not ret:
        break
    model=cv2.CascadeClassifier(face)
    facepoint=model.detectMultiScale(frame)
    for x,y,w,h in facepoint:
        cv2.rectangle(frame ,(x,y),(x+w,y+h),(0,0,255),2)

    cv2.imshow("face detection",frame)
    if cv2.waitKey(1) == ord('q'):#stop key is q 
        break
image_rgb=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.title("captured img")
plt.show()
cv2.destroyAllWindows()
cap.release()
```
